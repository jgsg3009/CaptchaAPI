```python
import matplotlib.pyplot as plt
import numpy as np
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Activation
from data_utils import *

%matplotlib inline
plt.rcParams['figure.figsize'] = (10.0, 8.0) # set default size of plots
plt.rcParams['image.interpolation'] = 'nearest'
plt.rcParams['image.cmap'] = 'gray'

import tensorflow as tf 
from tensorflow.keras import backend as k
import os
'''
config = tf.ConfigProto()
# config.gpu_options.per_process_gpu_memory_fraction = 0.1
config.gpu_options.allow_growth = True 
k.set_session(tf.Session(config=config))
'''
%load_ext autoreload
%autoreload 2
```

```python
from glob import glob
import cv2
import tensorflow.keras as keras

num_classes = 10

set_number = 1

train_len = 5
val_len = 1

imagePaths = glob('../pix2pix/predict/convertDataset/*/*')
val_imagePaths = glob('../pix2pix/predict/convertCourt/*/*')
x_train = []
y_train = []
x_val = []
y_val = []
for digit in range(6) :
    y_train.append([])
    y_val.append([])
for path in imagePaths :
    x_train.append(cv2.imread(path,cv2.IMREAD_UNCHANGED))
    for digit in range(6) :
        y_train[digit].append(int(path[42:-4][digit]))
for path in val_imagePaths :
    x_val.append(cv2.imread(path,cv2.IMREAD_UNCHANGED))
    for digit in range(6) :
        y_val[digit].append(int(path[40:-4][digit]))        
        
x_train = np.array(x_train)
x_val = np.array(x_val)
y_train = list(map(lambda y : keras.utils.to_categorical(y,10),y_train))
y_val = list(map(lambda y : keras.utils.to_categorical(y,10),y_val))
print(x_train.shape)
print(x_val.shape)
```

```python
from classifiers.ResNet_Captcha import ResNet18_6_Parted_NoPooling as ResNet18
from tensorflow.keras import losses
import tensorflow

weight_decay = 5e-4
lr = 1e-2
num_classes = 10
inceptionV1 = ResNet18(input_shape=(40, 120, 4), classes=num_classes, weight_decay=weight_decay)

#opt = tf.keras.optimizers.SGD(lr=lr, momentum=0.9, nesterov=False)
opt = tf.keras.optimizers.Adam(lr=lr,epsilon=0.1, beta_1=0.9, beta_2=0.999, amsgrad=True)
#opt = tf.keras.optimizers.RMSprop(lr=lr, rho=0.9, epsilon=0.1, decay=0.0)
inceptionV1.compile(optimizer=opt,
                    loss=losses.categorical_crossentropy,
                    metrics=['accuracy'])
inceptionV1.load_weights('./train_weight/Resnet_Captcha_1ì°¨.h5')
#inceptionV1.summary()
```

```python
%%time
from keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import ReduceLROnPlateau
from tensorflow.keras.callbacks import LearningRateScheduler
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.callbacks import CSVLogger
from lsuv_init import LSUVinit

# fit data with data augmentation or not
data_augmentation = False
lsuv_init = False
lr_schedule = [100,200,300]
batch_size = 100
epochs = lr_schedule[-1]

def lr_scheduler(epoch):        
    
    if epoch == 1 :
        return lr
    else : 
        return lr * (0.1 ** ((epoch-1)//40))

def lr_scheduler(epoch):
    global lr_schedule

    for i in range(0,len(lr_schedule)) :
        step = i
        if epoch < lr_schedule[i] :
            break;
    new_lr = lr * (0.25 ** (step))
    print(step)
    return new_lr

reduce_lr = LearningRateScheduler(lr_scheduler)
# reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.1,
#                               patience=10, min_lr=1e-6, verbose=1)
save_c = ModelCheckpoint('Resnet_Captcha_real_best_acc.h5', monitor="val_loss", save_best_only=True)
es = EarlyStopping(monitor='loss', mode='min', verbose=1, patience=2)
csv_logger = CSVLogger('log.csv', append=False, separator=';')

if lsuv_init :
    inceptionV1 = LSUVinit(inceptionV1,x_train[:batch_size])
    inceptionV1.save_weights('lusv_init.h5')

if data_augmentation:
    # datagen
    datagen = ImageDataGenerator(
        featurewise_center=False,  # set input mean to 0 over the dataset
        samplewise_center=False,  # set each sample mean to 0
        featurewise_std_normalization=False,  # divide inputs by std of the dataset
        samplewise_std_normalization=False,  # divide each input by its std
        zca_whitening=False,  # apply ZCA whitening
        rotation_range=5,  # randomly rotate images in the range (degrees, 0 to 180)
        width_shift_range=0.1,  # randomly shift images horizontally (fraction of total width)
        height_shift_range=0.1,  # randomly shift images vertically (fraction of total height)
        horizontal_flip=0,  # randomly flip images
        vertical_flip=False, # randomly flip images
    ) 
    # (std, mean, and principal components if ZCA whitening is applied).
    # datagen.fit(x_train)
    print('train with data augmentation')
    y_train = np.transpose(y_train,(1,0,2)).tolist()
    history = inceptionV1.fit_generator(generator=datagen.flow(x_train, y_train, batch_size=batch_size),
                                        epochs=epochs,
                                        callbacks=[reduce_lr,save_c,csv_logger]
                                        ,validation_data=(x_val, y_val)
                                        )
else:
    print('train without data augmentation')
    history = inceptionV1.fit(x_train, y_train, 
                  batch_size=batch_size, epochs=epochs, 
                  callbacks=[reduce_lr,save_c,csv_logger]
                  ,validation_data=(x_val, y_val)
                  )
    '''
    for i in range(0,20) :
        print(i) 
        history = inceptionV1.fit(x_train, y_train, 
                          batch_size=batch_size, epochs=epochs, 
                          callbacks=[reduce_lr,save_c,es],
                          validation_data=(x_val, y_val)
                          )
    '''                      
    
```

```python
# epoch 120 with adam,LUSV
# plot loss and acc 
plt.figure(figsize=(16, 12))
plt.subplot(len(lr_schedule),1,1)
plt.plot(history.history['loss'])
plt.xlabel('epoch')
plt.ylabel('Loss value')
plt.legend(['train', 'test'], loc='upper left')

for index in range(2,len(lr_schedule)+1) :
    plt.figure(figsize=(16, 12))
    plt.subplot(len(lr_schedule),1,index)
    plt.plot(history.history['loss'][lr_schedule[index-2],lr_schedule][index-1])
    plt.xlabel('epoch')
    plt.ylabel('Loss value')
    plt.legend(['train', 'test'], loc='upper left')    
    
    

plt.show()
```
